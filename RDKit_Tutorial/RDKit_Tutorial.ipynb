```python
# SMILES to mol
smiles = 'O=C(c1cc(=O)[nH]c2ccccc12)N1CCN(c2ccccc2)CC1'
m1 = Chem.MolFromSmiles(smiles)
m1
```

```python
# save image files and mol files
MolToFile(m1,'m1_plain.svg')
Chem.MolToMolFile(m1,'m1.mol')
```

```python
# reading from mol files
m1 = Chem.MolFromMolFile('m1.mol')
smiles = Chem.MolToSmiles(m1)
print(smiles)
m1 = Chem.MolFromSmiles(smiles)
m1
```

```python
# render customized vector images
drawsize = [400,200]
d = rdMolDraw2D.MolDraw2DSVG(drawsize[0], drawsize[1])
d.DrawMolecule(m1)
d.FinishDrawing()
svg = d.GetDrawingText()
display(SVG(svg.replace('svg:','')))
```

```python
# add notations and highlights
d = rdMolDraw2D.MolDraw2DSVG(drawsize[0], drawsize[1])
m1.GetAtomWithIdx(2).SetProp('atomNote', 'carbon')
m1.GetBondWithIdx(0).SetProp('bondNote', 'double')
d.drawOptions().addStereoAnnotation = True
d.drawOptions().addAtomIndices = True
d.DrawMolecule(m1)
d.FinishDrawing()
svg = d.GetDrawingText()
display(SVG(svg.replace('svg:','')))
```

```python
m1 = Chem.MolFromSmiles(smiles)
pattern = Chem.MolFromSmiles('c1ccccc1')
aromatic_atoms = list(m1.GetSubstructMatches(pattern))
aromatic_atoms
```

```python
aromatic_bonds = []
for aromatic_atoms_group in aromatic_atoms:
    for bond in pattern.GetBonds():
        atom_id1 = aromatic_atoms_group[bond.GetBeginAtomIdx()]
        atom_id2 = aromatic_atoms_group[bond.GetEndAtomIdx()]
        aromatic_bonds.append(m1.GetBondBetweenAtoms(atom_id1,atom_id2).GetIdx())
aromatic_bonds
```

```python
aromatic_atoms = [item for tup in aromatic_atoms for item in tup]
aromatic_atoms
```

```python
d = rdMolDraw2D.MolDraw2DSVG(drawsize[0], drawsize[1])
rdMolDraw2D.PrepareAndDrawMolecule(d, m1, highlightAtoms=aromatic_atoms,
                                   highlightBonds=aromatic_bonds)
d.FinishDrawing()
svg = d.GetDrawingText()
display(SVG(svg.replace('svg:','')))
```

```python
from matplotlib.colors import ColorConverter
atom_colors = {}
for i, atom in enumerate(aromatic_atoms):
    atom_colors[atom] = ColorConverter().to_rgb('tab:red')
atom_colors
```

```python
d = rdMolDraw2D.MolDraw2DSVG(drawsize[0], drawsize[1])
rdMolDraw2D.PrepareAndDrawMolecule(d, m1, highlightAtoms=aromatic_atoms,
                                   highlightBonds=aromatic_bonds,
                                   highlightAtomColors=atom_colors)
d.FinishDrawing()
svg = d.GetDrawingText()
display(SVG(svg.replace('svg:','')))
```

```python
m1 = Chem.MolFromSmiles(smiles)
atom_colors = {}
atom_colors_dict = {
    6: ColorConverter().to_rgb('tab:red'),
    7: ColorConverter().to_rgb('tab:orange'),
    8: ColorConverter().to_rgb('tab:cyan')
}
all_atoms = []
for atom in m1.GetAtoms():
    idx = atom.GetIdx()
    all_atoms.append(idx)
    atom_colors[idx] = atom_colors_dict[atom.GetAtomicNum()]

bond_colors = {}
bond_colors_dict = {
    float(1.0): ColorConverter().to_rgb('lightgray'),
    float(1.5): ColorConverter().to_rgb('tab:red'),
    float(2.0): ColorConverter().to_rgb('tab:cyan'),
}
all_bonds = []
for bond in m1.GetBonds():
    idx = bond.GetIdx()
    all_bonds.append(idx)
    bond_colors[idx] = bond_colors_dict[bond.GetBondTypeAsDouble()]
```

```python
d = rdMolDraw2D.MolDraw2DSVG(drawsize[0], drawsize[1])
rdMolDraw2D.PrepareAndDrawMolecule(d, m1, highlightAtoms=all_atoms,
                                   highlightBonds=all_bonds,
                                   highlightAtomColors=atom_colors,
                                   highlightBondColors=bond_colors)
d.FinishDrawing()
svg = d.GetDrawingText()
display(SVG(svg.replace('svg:','')))
```

```python
with open('m1_fancy.svg','w') as f:
    f.write(svg)
```

## The Goods: Molecular Descriptors

```python
from rdkit import Chem
from rdkit.Chem import Descriptors,AllChem,rdFreeSASA
```

```python
smiles = 'O=C(c1cc(=O)[nH]c2ccccc12)N1CCN(c2ccccc2)CC1'
m = Chem.MolFromSmiles(smiles)
```

```python
# descriptors
print(Descriptors.MolWt(m))
print(Descriptors.MolLogP(m))
print(Descriptors.NumHDonors(m))
print(Descriptors.NumHAcceptors(m))
print(Descriptors.TPSA(m,includeSandP=True))
AllChem.EmbedMolecule(m)
radii = rdFreeSASA.classifyAtoms(m)
print(rdFreeSASA.CalcSASA(m,radii))
```

```python
# fingerprints
m_fp = np.array(AllChem.GetMorganFingerprintAsBitVect(m, 3, nBits=1024))
m_fp.shape
m_fp
```

```python
# function to get the similarity values
def get_similarities(smiles_list, base_mol_fp, fpsize = 1024, check_smiles = False):
    
    # initialize similarity metrics
    sim_cosine = []
    sim_dice = []
    sim_euclidean = []

    for idx, smiles in enumerate(smiles_list):
        m = Chem.MolFromSmiles(smiles)

        # get the fingerprints for the molecule to compare
        m_fp = np.array(AllChem.GetMorganFingerprintAsBitVect(m, 3, nBits=1024))

        # calculate basic components
        both_AB = np.sum(base_mol_fp*m_fp)
        sum_A = np.sum(base_mol_fp)
        sum_B = np.sum(m_fp)

        #calculate similarities
        sim_cosine.append(both_AB / np.sqrt(sum_A*sum_B))
        sim_dice.append(2*both_AB / (sum_A + sum_B))
        sim_euclidean.append(np.sqrt((fpsize-sum_A-sum_B+2*both_AB) / fpsize))
            
    return sim_cosine, sim_dice, sim_euclidean

# test on an example
mtarget = Chem.MolFromSmiles('O=C(c1cc(=O)[nH]c2ccccc12)N1CCN(c2ccccc2)CC1')
mtarget_fp = np.array(AllChem.GetMorganFingerprintAsBitVect(mtarget,3,nBits=1024))
test_smiles = ['O=N(C1=CC=C(C2CCC(C(C3C4=C(C=CC=C4)NC(=O)C=3)=O)CC2)C=C1)=O']
get_similarities(test_smiles, mtarget_fp)
```

```python
from rdkit import Chem
from rdkit.utils import chemdraw
from rdkit.Chem.EnumerateStereoisomers import EnumerateStereoisomers, StereoEnumerationOptions
```

```python
chemdraw.StartChemDraw() # visible=False can cause problems
```

```python
smiles = 'CC=CC'
m1 = Chem.MolFromSmiles(smiles)
isomers = tuple(EnumerateStereoisomers(m1))
print('# of stereoisomers: {}'.format(len(isomers)))
for idx,smile in enumerate(sorted(Chem.MolToSmiles(x, isomericSmiles=True) for x in isomers)):
    print(smile)
    m1_cd = chemdraw.CDXConvert(smile,'chemical/x-daylight-smiles', 'image/png')
    with open('m1_stereo{}_cd.png'.format(idx),'wb') as f:
        f.write(m1_cd.tobytes())
    m1_cd = chemdraw.CDXConvert(smile,'chemical/x-daylight-smiles', 'chemical/cdx')
    with open('m1_stereo{}_cd.cdx'.format(idx),'wb') as f:
        f.write(m1_cd.tobytes())
```
